# Model Building

We'll start by importing the necessary functions from our source code, then load and clean the data.


```python
# Import the functions from the source code
from data_preprocessing import load_and_clean_data, preprocess_group_df, filter_data
from model_training import train_models, split_data
from data_visualization import plot_actual_vs_predicted

# Load and clean the data
df = load_and_clean_data('../data/property_data.csv')

# Print the first few rows of the dataframe to verify the data was loaded correctly
df.head()
```

Now, let's proceed with the main analysis, where we train the models and generate plots for each region.


```python
for (property_type, region), group_df in df.groupby(['Type of property', 'Region']):
    group_df = filter_data(group_df, property_type, region)
    group_df = preprocess_group_df(group_df, property_type)

    # Assuming 'Price of property in euro' is the target variable
    X = group_df.drop(['Price of property in euro', 'Type of property', 'Region'], axis=1)
    y = group_df['Price of property in euro']

    X_train, X_test, y_train, y_test = split_data(X, y)

    train_models(X_train, X_test, y_train, y_test, property_type, region)

    X = group_df[['Living area']]
    plot_actual_vs_predicted(X, y, property_type, region)
```
